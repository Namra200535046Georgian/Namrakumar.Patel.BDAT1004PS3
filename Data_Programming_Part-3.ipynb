## Question 1
#### 1.1 Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

#### 1.2 Import the dataset & Assign it to a variable called users

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep='|')
users
```

#### 1.3 Discover what is the mean age per occupation

```python
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
mean_age_per_occupation
```

#### 1.4 Discover the Male ratio per occupation and sort it from the most to the least 

```python
# Calculate the male ratio per occupation
male_ratio_per_occupation = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').sum() / x.count())

# Sort the male ratio from the most to the least
sorted_male_ratio = male_ratio_per_occupation.sort_values(ascending=False)

print(sorted_male_ratio)
```

#### 1.5 For each occupation, calculate the minimum and maximum ages

```python
min_max_age_occupation = users.groupby('occupation')['age'].agg([('min_age', 'min'), ('max_age', 'max')])
print(min_max_age_occupation)
```

#### 1.6 For each combination of occupation and sex, calculate the mean age

```python
age_stats_per_occupation = users.groupby(['occupation', 'gender'])['age'].mean()

print(age_stats_per_occupation)
```

#### 1.7 For each occupation present the percentage of women and men

```python
percentage_men_women = pd.crosstab(users['occupation'], users['gender'], normalize='index') * 100
print(percentage_men_women)
```

## Question 6
> Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend. 

```python
import pandas as pd
import matplotlib.pyplot as plt
df_6 = pd.read_csv('us-marriages-divorces-1867-2014.csv')
df_6.head()
```

```python
# Set the year column as the index
df_6.set_index('Year', inplace=True)

# Create a line plot showing the number of marriages and divorces per capita
plt.plot(df_6['Marriages_per_1000'], label='Marriages')
plt.plot(df_6['Divorces_per_1000'], label='Divorces')

# Add labels and legend
plt.title('Marriages and Divorces per Capita in the U.S. between 1867 and 2014')
plt.xlabel('Year')
plt.ylabel('Number per 1000')
plt.legend()

# Show the plot
plt.show()
```

## Question 7
> Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.


```python
import pandas as pd
import matplotlib.pyplot as plt
df_7 = pd.read_csv('us-marriages-divorces-1867-2014.csv')
df_7.head()
```

```python
years = [1900, 1950, 2000]
df_7 = df_7[df_7.Year.isin(years)]


fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(df_7.Year, df_7.Marriages_per_1000, width=3, align='edge', label='Marriages per capita')
ax.bar(df_7.Year, df_7.Divorces_per_1000, width=-3, align='edge', label='Divorces per capita')


ax.set_xlabel('Year')
ax.set_ylabel('Per capita')
ax.set_title('Marriages and Divorces per capita in the U.S. between 1900, 1950, and 2000')
ax.legend()
plt.show()
```

## Question 9
> Create a pie chart showing the fraction of all Roman Emperors that were assassinated.

> Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.

```python
df_9 = pd.read_csv("roman-emperor-reigns.csv")
df_9
```

```python
# the count of emperors in each cause of death category
cause_counts = df_9['Cause_of_Death'].value_counts()

# plotting a pie chart
plt.figure(figsize=(8, 8))
plt.pie(cause_counts, labels=cause_counts.index, autopct='%1.1f%%', startangle=140)

# title of the plot
plt.title('Fraction of Roman Emperors by Cause of Death')

# Equal aspect ratio ensures that the pie chart is circular
plt.axis('equal')

# Displaying the plot
plt.show()
```
